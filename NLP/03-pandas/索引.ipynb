# Pandas 索引结构

```python
import pandas as pd
df = pd.read_csv('./data/titanic.csv')
```

```python
df['Age'][:5] #定位一个列
```

### 定位两个列

```python
df[['Age','Fare']][:5]
```

### 定位某一个值
#### 用索引值定位
+ loc 用label来定位
+ iloc 用position来定位 

```python
df.iloc[0:5]
```

```python
df.iloc[0:5,1:3] #行，列
```

```python
df = df.set_index('Name')
```

```python
df.loc['Heikkinen, Miss. Laina','Fare']
```

```python
df.loc['Heikkinen, Miss. Laina':'Allen, Mr. William Henry']
```

### 改数据

```python
df.loc['Heikkinen, Miss. Laina','Fare'] = 1000
df.head()
```

### boolean 类型的索引

```python
df['Fare'] > 40
```

```python
df[df['Fare'] > 40]
```

## 综合处理

```python
df[df['Sex'] == 'female'][:5]
```

```python
df.loc[df['Sex'] == 'female','Age'].mean()
```

```python
(df['Age']>70).sum()
```
